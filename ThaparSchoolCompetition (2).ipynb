```python
x = train.drop(columns=['id','yield'],axis=1)
y = train['yield']
X_test = test.drop('id',axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
import xgboost as xgb
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
pipe = Pipeline(steps=[
    ('scaler',StandardScaler())
])

transform = ColumnTransformer(transformers=[
    ('col_transform',pipe)
],
remainder='passthrough'   
)
```

```python
kfold = KFold(n_splits=5,shuffle=True,random_state=42)
params = {
    'objective': 'reg:absoluteerror',
    'n_estimators': 2000,
    'learning_rate': 0.02,
    'max_depth': 5,
    'seed': 42,
    'n_jobs': -1,
    'subsample': 0.6,
    'reg_alpha': 0.5,
    'reg_lambda': 0.5,
    'min_child_weight' : 3,
    'early_stopping_rounds' : 160,
    'eval_metric' : 'mae'
}
ss = StandardScaler()

mae_score = []
test_pred = []
for train_idx,val_idx in kfold.split(x):
    x_train,x_test = x.iloc[train_idx],x.iloc[val_idx]
    y_train,y_test = y.iloc[train_idx],y.iloc[val_idx]
    x_train_scaled = ss.fit_transform(x_train)
    x_test_scaled = ss.transform(x_test)
    X_test_scaled = ss.transform(X_test)



    model = xgb.XGBRegressor(**params)
    model.fit(x_train_scaled,y_train,eval_set=[(x_test_scaled,y_test)],verbose=False)

    model_pred = model.predict(x_test_scaled)
    mae = mean_absolute_error(y_test,model_pred)
    test_model = model.predict(X_test_scaled)
    test_pred.append(test_model)
    mae_score.append(mae)

final_test_pred = np.mean(test_pred,axis=0)
print('Mean MAE Score : ',np.mean(mae_score))
```
